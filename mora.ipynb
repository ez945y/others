```python
import random
import os
```

```python
class Mora:
    def __init__(self,name,c_note=0,p_note=0):
        self.name=name
        self.lst=["       _   (^)\n      ( \   |_|\n       \_\  |_|\n       _\_\,/_|\n      (`\(_|`\|\n     (`\,)  \ \n      \,)   | |\n        \__(__|","   ,--.--._\n------\" _, \__)\n        / /___)\n        \//(__)\n------\     (_)\n       `-----\"","        _____\n .----   ___)____\n|    .-'    ____)\n|    ---    _____)\n|    `-.   _____)\n `---.___________)"]
        self.c_note, self.p_note=0,0
        self.computer=random.randint(1,3)
        self.rule="請閱讀遊戲規則:\n請選擇任意一個作為猜拳\n判出勝負後贏的猜輸的頭轉到哪裡\n如果你贏了選上下左右其中一個方向\n若你輸了就要被猜，猜中就結束此回合\n猜錯再繼續重複，先得兩分者獲勝"     
        print("歡迎",self.name,"猜拳")
        print(self.rule)
         
    def TopicInput(self):
        self.person=eval(input("[1]剪刀 [2]石頭 [3]布\n輸入你所要猜的拳:"))
        while self.person!=1 and self.person!=2 and self.person!=3 or self.person=='':
            self.person=eval(input("不符合格式，請再重新輸入:"))
        self.computer=random.randint(1,3)
        print()
        print(self.name,"出\n",self.lst[self.person-1])
        print("電腦出\n",self.lst[self.computer-1])   
    
    def Referee(self):
        if(self.computer == self.person):
            print("平手了，再猜一次\n")
            result="pease"
            return result
        elif(self.computer==2 and self.person==1) or (self.computer==1 and self.person==3) or (self.computer==3 and self.person==2):
            print("電腦贏了\n")
            result="lost"
            return result
        
        else:
            print(self.name,"贏了\n")
            result="win"
            return result
```

```python
class Guess(Mora):
    def __init__(self,name,c_note,p_note):
        super().__init__(name,c_note,p_note)
        self.face="   #############\n   ##         ##\n   #  ╭     ╮  #\n   #  ()   ()  #\n   (     ^     )\n    |         |\n    |   ╰–╯   |"
        self.faceL=["  #############\n   ##╭       ╮##\n   #  ()   ()  #\n   #     ^     #\n   (    ╰–╯    )\n    |         |\n    |         |","  ############# \n   ##         ## \n   # ╭       ╮ #\n   (           )\n    | ()   () |\n    |    ^    |\n    \   ╰–╯   /","  #############\n   ##         ##\n   # ╭     ╮   #\n   # ()   ()   #\n   (   ^       )\n    |         |\n    | ╰–╯     |","  #############\n   ##         ## \n   #  ╭       ╮#\n   #   ()   () #\n   (       ^   )\n    |         | \n    |      ╰–╯|"]
        self.hand=["↑","↓","←","→"]
        self.computer=random.randint(1,4)
        self.rule="[1]上 [2]下 [3]左 [4]右"
        
    def TopicInputW(self):
        print()
        print(self.face)
        self.person=eval(input("你選擇你所要比的方向:"))
        while self.person!=1 and self.person!=2 and self.person!=3 and self.person!=4 or self.person=='':
            self.person=eval(input("不符合格式，請再重新輸入:"))
        self.computer=random.randint(1,4)
        print("\n你指向:\n",self.hand[self.person-1])
        print("\n電腦轉向:\n",self.faceL[self.computer-1])
        self.RefereeW()
        
    def TopicInputL(self):
        print()
        self.person=eval(input("你選擇你所要轉頭的方向:"))
        while self.person!=1 and self.person!=2 and self.person!=3 and self.person!=4:
            self.person=eval(input("不符合格式，請再重新輸入:"))
        self.computer=random.randint(1,4)
        print("\n電腦比向:\n",self.hand[self.computer-1])
        print("\n你轉向:\n",self.faceL[self.person-1])
        self.RefereeL()
        
    def RefereeL(self):
        if(self.computer == self.person):
            print("被猜中了\n")
            self.c_note+=1
            
        else:
            print("你成功躲過了!\n")
            
                
    def RefereeW(self):
        if(self.computer == self.person):
            print("你猜對了!\n")
            self.p_note+=1
            
        else:
            print("可惜再來一次\n")
```

```python
def main():
    name=input("歡迎參加黑白切比賽！請輸入你的名字:")
    mora=Mora(name)
    guess=Guess(mora.name,mora.c_note,mora.p_note)

    while guess.c_note<2 and guess.p_note<2:
        mora.TopicInput()
        result=mora.Referee()
        if result=="pease":
            continue
        elif result=="win":
            print(guess.rule)
            guess.TopicInputW()
        else:
            print(guess.rule)
            guess.TopicInputL()
        print("分數\n電腦:",guess.c_note,mora.name,":",guess.p_note,"\n")

    if guess.p_note==2:
        print(guess.name,"獲勝")
    else:
        print("電腦獲勝")
main()
```
